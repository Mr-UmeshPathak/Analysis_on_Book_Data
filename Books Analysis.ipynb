```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('books.csv', error_bad_lines = False)
books[:10]
```

```python
#Since we don't need isbn,isbn13 and bookID for analysis purpose, hence dropping the columns

books = books.drop(columns=['isbn', 'isbn13','bookID'])
books.head()
```

```python
books.columns
```

```python
books.describe()
```

```python
#We see below that there are no null or missing values, therefore we don't need to fill in any missing values

books.info()
```

```python
#Let's try to get the top rated books by average rating. "Top Rated" means best rated by readers.

top_rated_books = books.sort_values(by = ['average_rating'], ascending = False)
top_rated_books
```

```python
"""
We can observe in above data that though the ratings are 5 or so, 
but few people have actually rated it. This gives us inaccurate data, 
since we need books which have been rated as top books by more readers. 
Therefore, we next to try to take top rated books where it has been rated at least by 100 readers. """
```

```python
top_rated_books = [x for x in books['ratings_count'] > 100]
```

```python
top_rated_books = books[top_rated_books]
```

```python
top_rated_books = top_rated_books.sort_values(by = ['average_rating'], ascending = False)
```

```python
top_rated_books = top_rated_books[:100]
```

```python
top_rated_books
```

```python
top_rated_books.describe()
```

```python
#Below are the authors and publsihers of top rated books
```

```python
top_rated_authors = top_rated_books["authors"].value_counts()[:20]
top_rated_authors.head()
```

```python
top_rated_publishers = top_rated_books["publisher"].value_counts()[:20]
top_rated_publishers
```

```python
#Now that we have books which are best rated, 
#let's find out books which might not have the best rating but our loved by most readers. 
#We call it the "most popular" category.

most_popular_books = books.sort_values(by=['ratings_count','text_reviews_count'], ascending=False)[:20]
most_popular_books
```

```python
most_popular_hundred = books.sort_values(by=['ratings_count','text_reviews_count'], ascending=False)[:100]
```

```python
most_popular_authors = most_popular_hundred["authors"].value_counts()[:20]
most_popular_authors
```

```python
publishers_with_popular_books = most_popular_hundred["publisher"].value_counts()[:20]
publishers_with_popular_books
```

```python
#Let's find out how many books authors and publishers have published. 
#Also,The maximum and minimum number of published books by them.

publishers_most_books = books["publisher"].value_counts()
publishers_most_books.head(20)
```

```python
#Maximum number of books by a publisher
publishers_most_books.max()
```

```python
#Minimum number of books by publisher
publishers_most_books.min()
```

```python
author_most_books = books["authors"].value_counts()
author_most_books.head(20)
```

```python
#Maximum number of books by an author
author_most_books.max()
```

```python
#Minimum number of books by an author
author_most_books.min()
```

```python
#We find the mean number of books by publishers and the authors
#Average books by a publisher

round(publishers_most_books.mean(),2)
```

```python
#average books by an author
round(author_most_books.mean(),2)
```
